```python
genre_list = []
for genres in df_t.genre_ids:
    for genre in genres:
        if genre not in genre_list:
            genre_list.append(genre)
```

```python
for genre in genre_list:
    df_t[genre] = df_t['genre_ids'].apply(lambda x: genre in x)
    
df_t.head()
```
