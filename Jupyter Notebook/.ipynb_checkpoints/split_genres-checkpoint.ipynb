```python
df_t = pd.read_csv('df_100.csv')
```

```python
df_t.head()
```

```python
"""
This code give each genre a column in the dataframe.

"""
df_t.genre_ids = df_t.genre_ids.apply(lambda x: eval(x)) 
```

```python
"""
Below creates list that contains genres

"""

genre_list = []
for genres in df_t.genre_ids:
    for genre in genres:
        if genre not in genre_list:
            genre_list.append(genre)
```

```python
"""
Following code places 'True' or 'False' values in the genre columns taken from the 'genreids' column.

"""
for genre in genre_list:
    df_t[genre] = df_t['genre_ids'].apply(lambda x: genre in x)
    
print(genre_list)
```
